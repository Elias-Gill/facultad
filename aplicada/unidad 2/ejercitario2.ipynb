**Ejercicio 6.** Sea $L(x)=R(x)=x^a$, con $a>0$. Escribe la función de membresía y los cortes alfa en forma intervalar de los números L-R basados en estas funciones.

## Respuesta

### (i) Función de membresía:
$
u(x) =
\begin{cases} 
0, & \text{si } x < a_0^-, \\
\left( \frac{x - a_0^-}{a} \right)^q, & \text{si } a_0^- \leq x < a_1^-, \\
1, & \text{si } a_1^- \leq x < a_1^+, \\
\left( \frac{a_0^+ - x}{a} \right)^q, & \text{si } a_1^+ \leq x < a_0^+, \\
0, & \text{si } a_0^+ \leq x.
\end{cases}
$

### (ii) Los cortes alfa en forma intervalar de los números L-R basados en estas funciones:

$
\alpha = L\left[\left( \frac{x - a_0^-}{a} \right)^q\right] \Rightarrow \left( \frac{x - a_0^-}{a} \right)^q = L^{-1}(\alpha) \Rightarrow \left( \frac{x - a_0^-}{a} \right) = (\alpha)^\frac{1}{q} \Rightarrow x = a_0^- + (\alpha)^\frac{1}{q}a
$

$
\alpha = R\left[\left( \frac{a_0^+ - x}{a} \right)^q\right] \Rightarrow \left( \frac{a_0^+ - x}{a} \right)^q = R^{-1}(\alpha) \Rightarrow \left( \frac{a_0^+ - x}{a} \right) = (\alpha)^\frac{1}{q} \Rightarrow x = a_0^+ - (\alpha)^\frac{1}{q}a
$

Finalmente, los cortes alfa están dados por:
$
u_\alpha = [a_0^- + (\alpha)^\frac{1}{q}a, a_0^+ - (\alpha)^\frac{1}{q}a]
$

**Ejercicio 8.** Demuestra las propiedades de abajo.

1. Para cualquier número difuso $u$ y cualquier $a,b\in \mathbb{R}$, con $a\cdot b\geq 0$, se cumple $(a+b)\cdot u=(a\cdot u)+(b\cdot u)$.
2. Para cualquier $\lambda\in\mathbb{R}$ y cualesquiera números difusos $u,v$, se cumple $\lambda\cdot (u+v)=\lambda\cdot u+\lambda\cdot v$.
3. Para cualquier $\lambda,\mu\in\mathbb{R}$ y cualquier número difuso $u$, se cumple $(\lambda\cdot \mu)\cdot u=\lambda\cdot (\mu\cdot u)$.

## Respuestas:

La suma de dos números difusos $u + v$ y la multiplicación entre un real y un número difuso $\lambda \cdot u$ podemos definirlas como:

#### (i) $(u + v)(x) = u(x) + v(x)$.

Por ejemplo, si $u(x) = 4$, $a = 2$ y $b = 3$, entonces:

#### (ii) $(a + b) \cdot u(x) = a \cdot u(x) + b \cdot u(x) = (a \cdot u)(x) + (b \cdot u)(x)$.

Por ejemplo, si $u(x) = 4$, $a = 2$ y $b = 3$, entonces:

$
(2 + 3) \cdot 4 = 5 \cdot 4 = 20
$
$
(2 \cdot 4) + (3 \cdot 4) = 8 + 12 = 20
$

#### (iii) $\lambda \cdot (u + v) = \lambda \cdot u + \lambda \cdot v$.

Por ejemplo, si $u(x) = 3$, $v(x) = 4$, y $\lambda = 2$, entonces:

$
\lambda \cdot (u + v)(x) = (\lambda \cdot (u + v))(x) = \lambda \cdot u(x) + \lambda \cdot v(x)
$

$
2 \cdot (3 + 4) = 2 \cdot 7 = 14
$

$
(2 \cdot 3) + (2 \cdot 4) = 6 + 8 = 14
$

#### (iv) $(\lambda \cdot \mu) \cdot u = \lambda \cdot (\mu \cdot u)$.

Por ejemplo, si $u(x) = 4$, $\lambda = 2$, y $\mu = 3$, entonces:

$
(\lambda \cdot \mu) \cdot u(x) = ((\lambda \cdot \mu) \cdot u)(x) = (\lambda \cdot (\mu \cdot u))(x) = \lambda \cdot (\mu \cdot u)(x)
$

$
(2 \cdot 3) \cdot 4 = 6 \cdot 4 = 24
$

$
2 \cdot (3 \cdot 4) = 2 \cdot 12 = 24
$

## Respuestas:

(i) Escribe la función de membresía o el intervalo del corte alfa de $ u \cdot v $.

La multiplicación $ u \cdot v $ se define como:

$
u \cdot v = \text{trfn}(2, 5, 6, 8) \cdot \text{trfn}(-3, 1, 1, 6)
$

Reemplazando por los valores de $ u $ y $ v $ tenemos que:

$
u \cdot v = \text{trfn}(2, 5, 6, 8) \cdot \text{trfn}(-3, 1, 1, 6)
$

Finalmente, $ u \cdot v $ se puede representar como un número difuso trapezoidal por el cuádruple $ (a, b, c, d) $.

La función de membresía de $ u \cdot v $ es igual a:

$
u \cdot v = \text{trfn}(-24, 5, 6, 48)
$

El intervalo del corte alfa de $ u \cdot v $ se define como:

$
\begin{align}
a &= \min(2 \cdot -3, 2 \cdot 6, 8 \cdot -3, 8 \cdot 6) = \min(-6, 12, -24, 48) = -24 \\
b &= \min(5 \cdot 1, 5 \cdot 1, 6 \cdot 1, 6 \cdot 1) = \min(5, 5, 6, 6) = 5 \\
c &= \max(5 \cdot 1, 5 \cdot 1, 6 \cdot 1, 6 \cdot 1) = \max(5, 5, 6, 6) = 6 \\
d &= \max(2 \cdot -3, 2 \cdot 6, 8 \cdot -3, 8 \cdot 6) = \max(-6, 12, -24, 48) = 48
\end{align}
$

Por lo tanto, $ u \cdot v = \text{trfn}(-24, 5, 6, 48) $ donde $ (a, b, c, d) \in \mathbb{R}^4 $ y $ a \leq b \leq c \leq d $.


$
u(x) =
\begin{cases}
0, & \text{si } x < a \\
\frac{x - a}{b - a}, & \text{si } a \leq x < b \\
1, & \text{si } b \leq x < c \\
\frac{d - x}{d - c}, & \text{si } c \leq x < d \\
0, & \text{si } d \leq x
\end{cases}
$

La función de membresía $ u \cdot v = \text{trfn}(-24, 5, 6, 48) $ es igual a:

$
u \cdot v(x) =
\begin{cases}
0, & \text{si } x < -24 \\
\frac{x + 24}{29}, & \text{si } -24 \leq x < 5 \\
1, & \text{si } 5 \leq x < 6 \\
\frac{48 - x}{42}, & \text{si } 6 \leq x < 48 \\
0, & \text{si } 48 \leq x
\end{cases}
$

El intervalo de corte alfa se define como:

$
\alpha = \frac{x+24}{29} \Rightarrow x = 29 \cdot \alpha - 24
$

$
\alpha = \frac{48-x}{42} \Rightarrow x = 48 - 42 \cdot \alpha
$

La expresión final para $ u \cdot v $ es:

$
u \cdot v = [29 \cdot \alpha - 24, 48 - 42 \cdot \alpha]
$

**Ejercicio 10.** Escribe una expresión para los cortes alfa del producto de dos números difusos triangulares $u=\text{tfn}(a_1,b_1,c_1)$ y $v=\text{tfn}(a_2,b_2,c_2)$. Escribe la familia de cortes alfa en forma intervalar de $u\cdot v$.

## Respuesta

Sean $ u = \text{tfn}(a_1, b_1, c_1) $ y $ v = \text{tfn}(a_2, b_2, c_2) $ dos números difusos triangulares. Los cortes alfa de $ u $ y $ v $ se definen como:

### Producto Difuso

El producto difuso $ u \cdot v $ se define como el producto de los cortes alfa, es decir:
$
(u \cdot v)_\alpha = u_\alpha \cdot v_\alpha
$

### Familia de Cortes Alfa en Forma Intervalar

La familia de cortes alfa para el producto $ u \cdot v $ se expresa como:

$
(u \cdot v)_\alpha = \begin{cases}
[a_1 \cdot a_2, a_1 \cdot b_2, a_1 \cdot c_2, b_1 \cdot a_2, b_1 \cdot b_2, b_1 \cdot c_2, c_1 \cdot a_2, c_1 \cdot b_2, c_1 \cdot c_2] & \text{si } \alpha \leq \frac{1}{2} \\
[\min(a_1 \cdot a_2, b_1 \cdot b_2), \max(b_1 \cdot c_2, c_1 \cdot b_2)] & \text{si } \frac{1}{2} < \alpha < 1 \\
[c_1 \cdot a_2, c_1 \cdot b_2, c_1 \cdot c_2] & \text{si } \alpha = 1
\end{cases}
$
