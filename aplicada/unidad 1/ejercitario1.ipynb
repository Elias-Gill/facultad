**Ejercicio 3.** Demostrar que $([0,1],\leq)$ es un retículo completo y distributivo.

**Demostración de que $([0,1], \leq)$ es un retículo completo y distributivo**

Para demostrar que $([0,1], \leq)$ es un retículo completo y distributivo, primero definimos lo que significa que un conjunto parcialmente ordenado sea un retículo completo y distributivo:

1. **Retículo completo**: Un poset $(P, \leq)$ es un retículo completo si para cualquier subconjunto $A \subseteq P$, existen el supremo (unión) $\sup A$ y el ínfimo (intersección) $\inf A$ en $P$.
2. **Retículo distributivo**: Un retículo $(P, \leq)$ es distributivo si para todos $x, y, z \in P$ se cumple que:
   $x \wedge (y \vee z) = (x \wedge y) \vee (x \wedge z)$
   y
   $x \vee (y \wedge z) = (x \vee y) \wedge (x \vee z)$

### Retículo Completo

Consideremos $([0,1], \leq)$, donde el conjunto $[0,1]$ es el intervalo cerrado de los números reales entre 0 y 1, y la relación de orden $\leq$ es la orden usual en los números reales.

- **Supremo (unión)**: Dado un subconjunto $A \subseteq [0,1]$, el supremo de $A$ es simplemente el máximo de $A$. Esto es porque $[0,1]$ es un intervalo cerrado y acotado, por lo que todo subconjunto $A$ tiene un máximo.
  
- **Ínfimo (intersección)**: Similarmente, el ínfimo de $A$ es el mínimo de $A$, ya que $[0,1]$ es cerrado y acotado.

Dado que para cualquier subconjunto $A \subseteq [0,1]$ existen $\sup A$ y $\inf A$ en $[0,1]$, concluimos que $([0,1], \leq)$ es un retículo completo.

### Retículo Distributivo

Para demostrar que $([0,1], \leq)$ es un retículo distributivo, debemos verificar que las operaciones de supremo e ínfimo distribuyen sobre cada una:

- **Operación $\wedge$ (mínimo)**: Dado que $\wedge$ en $([0,1], \leq)$ es simplemente el mínimo entre dos elementos, podemos verificar que:
  
  $x \wedge (y \vee z) = \min(x, \max(y, z)) = \max(\min(x, y), \min(x, z)) = (x \wedge y) \vee (x \wedge z)$
  
- **Operación $\vee$ (máximo)**: Análogamente, dado que $\vee$ es el máximo entre dos elementos, podemos verificar que:
  
  $x \vee (y \wedge z) = \max(x, \min(y, z)) = \min(\max(x, y), \max(x, z)) = (x \vee y) \wedge (x \vee z)$

Esto muestra que $([0,1], \leq)$ es distributivo.

### Conclusión

Dado que hemos demostrado que $([0,1], \leq)$ es un retículo completo (todo subconjunto tiene supremo e ínfimo) y distributivo (las operaciones $\wedge$ y $\vee$ distribuyen sobre sí mismas), concluimos que $([0,1], \leq)$ es un retículo completo y distributivo.

**Ejercicio 7.** Demuestra las propiedades 1 al 5, 7 y 8 en la Proposición 1.15 de libro de Bede (2013).

### 1. **Asociatividad (Associativity)**

**Propiedad:**
$$
A \cup (B \cup C) = (A \cup B) \cup C
$$
$$
A \cap (B \cap C) = (A \cap B) \cap C
$$

**Demostración:**
Para un conjunto difuso \(A\), el operador \(\cup\) se define como:
$$
\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x))
$$

Entonces:
$$
\mu_{A \cup (B \cup C)}(x) = \max(\mu_A(x), \max(\mu_B(x), \mu_C(x)))
$$
Por la propiedad asociativa del máximo:
$$
\max(\mu_A(x), \max(\mu_B(x), \mu_C(x))) = \max(\max(\mu_A(x), \mu_B(x)), \mu_C(x))
$$
$$
= \mu_{(A \cup B) \cup C}(x)
$$
Por lo tanto:
$$
A \cup (B \cup C) = (A \cup B) \cup C
$$

La misma lógica se aplica para \(\cap\) usando la operación \(\min\):
$$
\mu_{A \cap (B \cap C)}(x) = \min(\mu_A(x), \min(\mu_B(x), \mu_C(x))) = \min(\min(\mu_A(x), \mu_B(x)), \mu_C(x))
$$
$$
= \mu_{(A \cap B) \cap C}(x)
$$
Por lo tanto:
$$
A \cap (B \cap C) = (A \cap B) \cap C
$$

### 2. **Conmutatividad (Commutativity)**

**Propiedad:**
$$
A \cup B = B \cup A
$$
$$
A \cap B = B \cap A
$$

**Demostración:**
Por la definición de \(\cup\) (máximo):
$$
\mu_{A \cup B}(x) = \max(\mu_A(x), \mu_B(x))
$$
Y dado que \(\max(a, b) = \max(b, a)\), se tiene:
$$
\mu_{A \cup B}(x) = \max(\mu_B(x), \mu_A(x)) = \mu_{B \cup A}(x)
$$
Por lo tanto:
$$
A \cup B = B \cup A
$$

De manera similar para \(\cap\) (mínimo):
$$
\mu_{A \cap B}(x) = \min(\mu_A(x), \mu_B(x)) = \min(\mu_B(x), \mu_A(x)) = \mu_{B \cap A}(x)
$$
Por lo tanto:
$$
A \cap B = B \cap A
$$

### 3. **Identidad (Identity)**

**Propiedad:**
$$
A \cup \emptyset = A
$$
$$
A \cap X = A
$$

**Demostración:**

Para la unión con el conjunto vacío (\(\emptyset\)):
$$
\mu_{\emptyset}(x) = 0 \quad \text{para todo } x
$$
$$
\mu_{A \cup \emptyset}(x) = \max(\mu_A(x), 0) = \mu_A(x)
$$
Por lo tanto:
$$
A \cup \emptyset = A
$$

Para la intersección con el conjunto universal (\(X\)):
$$
\mu_X(x) = 1 \quad \text{para todo } x
$$
$$
\mu_{A \cap X}(x) = \min(\mu_A(x), 1) = \mu_A(x)
$$
Por lo tanto:
$$
A \cap X = A
$$

### 4. **Absorción por \(\emptyset\) y \(X\) (Absorption by \(\emptyset\) and \(X\))**

**Propiedad:**
$$
A \cup X = X
$$
$$
A \cap \emptyset = \emptyset
$$

**Demostración:**

Para la unión con el conjunto universal (\(X\)):
$$
\mu_{A \cup X}(x) = \max(\mu_A(x), 1) = 1 = \mu_X(x)
$$
Por lo tanto:
$$
A \cup X = X
$$

Para la intersección con el conjunto vacío (\(\emptyset\)):
$$
\mu_{A \cap \emptyset}(x) = \min(\mu_A(x), 0) = 0 = \mu_{\emptyset}(x)
$$
Por lo tanto:
$$
A \cap \emptyset = \emptyset
$$

### 5. **Idempotencia (Idempotence)**

**Propiedad:**
$$
A \cup A = A
$$
$$
A \cap A = A
$$

**Demostración:**

Para la unión:
$$
\mu_{A \cup A}(x) = \max(\mu_A(x), \mu_A(x)) = \mu_A(x)
$$
Por lo tanto:
$$
A \cup A = A
$$

Para la intersección:
$$
\mu_{A \cap A}(x) = \min(\mu_A(x), \mu_A(x)) = \mu_A(x)
$$
Por lo tanto:
$$
A \cap A = A
$$

### 6. **Distributividad (Distributivity)**

**Propiedad:**
$$
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)
$$
$$
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)
$$

**Demostración:**

Para la primera propiedad:
$$
\mu_{A \cap (B \cup C)}(x) = \min(\mu_A(x), \max(\mu_B(x), \mu_C(x)))
$$
Por la distributividad de \(\min\) sobre \(\max\):
$$
= \max(\min(\mu_A(x), \mu_B(x)), \min(\mu_A(x), \mu_C(x)))
$$
$$
= \mu_{(A \cap B) \cup (A \cap C)}(x)
$$
Por lo tanto:
$$
A \cap (B \cup C) = (A \cap B) \cup (A \cap C)
$$

Para la segunda propiedad:
$$
\mu_{A \cup (B \cap C)}(x) = \max(\mu_A(x), \min(\mu_B(x), \mu_C(x)))
$$
Por la distributividad de \(\max\) sobre \(\min\):
$$
= \min(\max(\mu_A(x), \mu_B(x)), \max(\mu_A(x), \mu_C(x)))
$$
$$
= \mu_{(A \cup B) \cap (A \cup C)}(x)
$$
Por lo tanto:
$$
A \cup (B \cap C) = (A \cup B) \cap (A \cup C)
$$

### 7. **Involución (Involution)**

**Propiedad:**
$$
\overline{\overline{A}} = A
$$

**Demostración:**

El complemento de un conjunto difuso \(A\) está dado por:
$$
\mu_{\overline{A}}(x) = 1 - \mu_A(x)
$$

Entonces, al tomar el complemento del complemento:
$$
\mu_{\overline{\overline{A}}}(x) = 1 - \mu_{\overline{A}}(x) = 1 - (1 - \mu_A(x)) = \mu_A(x)
$$
Por lo tanto:
$$
\overline{\overline{A}} = A
$$

2. **Asociatividad**: $S_L(x, S_L(y, z)) = S_L(S_L(x, y), z)$.
3. **Idempotencia**: $S_L(x, x) = x$.
4. **Monotonía**: Si $x \leq x'$ y $y \leq y'$, entonces $S_L(x, y) \leq S_L(x', y')$.

**Conmutatividad**:
$$
x S_L y = (x + y) \land 1 = y S_L x.
$$

**Asociatividad**:
Verificamos:
$$
x S_L (y S_L z) = x S_L ((y + z) \land 1) = (x + (y + z) \land 1) \land 1.
$$
$$
(x S_L y) S_L z = ((x + y) \land 1) S_L z = ((x + y) \land 1 + z) \land 1.
$$

Para mostrar que ambas expresiones son iguales:
$$
x S_L (y S_L z) = (x + y + z) \land 1
$$

$$
(x + y + z) \land 1
$$

**Idempotencia**:
$$
x S_L x = (x + x) \land 1 = 2x \land 1 = x.
$$

**Monotonía**:
Si $x \leq x'$ y $y \leq y'$, entonces:
$$
x S_L y = (x + y) \land 1 \leq (x' + y') \land 1 = x' S_L y'.
$$

Entonces, $S_L$ cumple con las propiedades de una t-conorma.

### Conclusión

Hemos demostrado que tanto $T_L$ como $S_L$ satisfacen las propiedades necesarias para ser una t-norma y una t-conorma, respectivamente.

**Ejercicio 11.** Considera los conjuntos difusos $A,B:\mathbb{R}^+\to[0,1]$ definidos como $A(x)=1/(1+x^2)$ y $B(x)=1/10^x$. Escribe formalmente como funciones $AT_LB$, $AS_LB$, $AT_GB$, $AS_GB$, donde $T_L$ y $S_L$ son la t-norma y t-conorma de Łukasiewicz y $T_G$ y $S_G$ son la t-norma y t-conorma de Goguen. Presenta una gráfica de cada uno.

```python
import numpy as np
import matplotlib.pyplot as plt


# Definición de los conjuntos difusos A y B
def A(x):
    return 1 / (1 + x**2)


def B(x):
    return 1 / (10**x)


# t-norma y t-conorma de Łukasiewicz
def T_Lukasiewicz(a, b):
    return max(0, a + b - 1)


def S_Lukasiewicz(a, b):
    return min(1, a + b)


# t-norma y t-conorma de Goguen
def T_Goguen(a, b):
    return a * b


def S_Goguen(a, b):
    return a + b - a * b


# Funciones para calcular AT_LB, AS_LB, AT_GB, AS_GB
def AT_LB(x):
    return T_Lukasiewicz(A(x), B(x))


def AS_LB(x):
    return S_Lukasiewicz(A(x), B(x))


def AT_GB(x):
    return T_Goguen(A(x), B(x))


def AS_GB(x):
    return S_Goguen(A(x), B(x))


# Rango de valores para graficar
x_values = np.linspace(0, 5, 100)
AT_LB_values = [AT_LB(x) for x in x_values]
AS_LB_values = [AS_LB(x) for x in x_values]
AT_GB_values = [AT_GB(x) for x in x_values]
AS_GB_values = [AS_GB(x) for x in x_values]

# Gráficas
plt.figure(figsize=(12, 8))

# Gráfica de AT_LB
plt.subplot(2, 2, 1)
plt.plot(x_values, AT_LB_values, label="AT_LB", color="blue")
plt.title("t-norma de Łukasiewicz (AT_LB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AS_LB
plt.subplot(2, 2, 2)
plt.plot(x_values, AS_LB_values, label="AS_LB", color="orange")
plt.title("t-conorma de Łukasiewicz (AS_LB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AT_GB
plt.subplot(2, 2, 3)
plt.plot(x_values, AT_GB_values, label="AT_GB", color="green")
plt.title("t-norma de Goguen (AT_GB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

# Gráfica de AS_GB
plt.subplot(2, 2, 4)
plt.plot(x_values, AS_GB_values, label="AS_GB", color="red")
plt.title("t-conorma de Goguen (AS_GB)")
plt.xlabel("x")
plt.ylabel("Valor")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
